```python
import math
```

```python
def PrimeChecker(a):   
    if a > 1:  
        for j in range(2, int(a/2) + 1):   
            if (a % j) == 0:  
                return False  
                break  
        else:  
            return True   
    else:  
        return False 
     
```

```python
def calculateDigits(num):
    
    noOfDgts=0
   
    while(num!=0):
        num=num/10
        num=math.floor(num)
        noOfDgts+=1
    return noOfDgts
```

```python
def truncation(num):
    
    count=calculateDigits(num)
    revCount=count*-1
    check1=0
    check2=1
    arr=str(num)
    val=0
    sum=0
    
    for j in range(count):
        val=int(arr[j:])
        if PrimeChecker(val)==True:
            check1+=1

    for j in range(-1,revCount,-1):
        val=int(arr[:j])
        if PrimeChecker(val)==True:
            check2+=1
    
    
    if check1==check2  and check1==count:
        print(num)
        return num
    else:
        return 0
    
```

```python
sum=0
for num in range(10,100000):
    sum=sum+truncation(num)

print("Sum of All Truncatable Primes below one million",":",sum)
```
